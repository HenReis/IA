## 1. Imports & Configuration

```python
# ── Silenciar avisos desnecessários do sklearn/matplotlib ──
import warnings, pickle
warnings.filterwarnings("ignore")

# Manipulação de arrays numéricos e tabelas de dados
import numpy as np
import pandas as pd

# Visualização de gráficos
import matplotlib.pyplot as plt
import seaborn as sns

# ── Utilitários de treino e avaliação do sklearn ──
from sklearn.model_selection import (
    train_test_split,             # divide os dados em treino e teste
    RandomizedSearchCV,           # pesquisa aleatória de hiperparâmetros
    cross_val_score,              # avaliação por validação cruzada
    StratifiedKFold,              # folds estratificados (mantém proporção de classes)
    KFold,                        # folds simples para regressão
    TunedThresholdClassifierCV    # wrapper sklearn nativo para optimizar o threshold de decisão
)

# Pipeline encadeia passos (pré-processamento + modelo) num único objecto
from sklearn.pipeline import Pipeline
# ColumnTransformer aplica transformações diferentes a colunas diferentes
from sklearn.compose import ColumnTransformer

# Pré-processamento
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer

# Métricas de avaliação
from sklearn.metrics import (
    r2_score,                  # R² para regressão
    mean_squared_error,        # MSE (usamos a raiz, RMSE)
    f1_score,                  # F1-Score para classificação
    classification_report,     # relatório completo precision/recall/F1
    ConfusionMatrixDisplay,    # visualização da matriz de confusão
    precision_recall_curve     # curva para encontrar threshold ótimo
)

# Algoritmos candidatos
from sklearn.linear_model import Ridge, LogisticRegression
from sklearn.ensemble import (
    RandomForestRegressor, GradientBoostingRegressor,
    RandomForestClassifier, GradientBoostingClassifier
)
from sklearn.svm import SVR, SVC

# Estilo global dos gráficos
sns.set_theme(style="whitegrid", palette="muted")

# Semente aleatória fixa — garante que todos os resultados são reproduzíveis
RANDOM_STATE = 42

print("Imports OK")
```

## 4. Feature Engineering & Pre-processing

```python
# ── Remoção de colunas sem valor preditivo ──
# EmployeeCount  : constante (sempre 1) — não transmite informação
# EmployeeNumber : identificador único — não é uma feature, é um índice
# Over18         : constante (todos "Y") — sem variabilidade
# StandardHours  : constante (sempre 80) — sem variabilidade
DROP_COLS = ["EmployeeCount", "EmployeeNumber", "Over18", "StandardHours"]
df = df.drop(columns=DROP_COLS)
print("Remaining columns:", df.shape[1])

# Nomes das colunas alvo (o que queremos prever)
TARGET_REG = "MonthlyIncome"  # tarefa de regressão
TARGET_CLF = "Attrition"      # tarefa de classificação

# ── Separação de features por tipo ──
# O pré-processador precisa de saber quais colunas são categóricas
# e quais são numéricas para aplicar transformações diferentes

# Colunas de texto/categóricas (dtype=object), excluindo o alvo de classificação
cat_cols = df.select_dtypes(include="object").columns.tolist()
cat_cols = [c for c in cat_cols if c != TARGET_CLF]

# Colunas numéricas (dtype=int64/float64), excluindo o alvo de regressão
num_cols = df.select_dtypes(include=np.number).columns.tolist()
num_cols = [c for c in num_cols if c != TARGET_REG]

print("Numeric features:", len(num_cols))
print("Categorical features:", len(cat_cols))
```

```python
# ── Pré-processador partilhado pelas duas pipelines ──
# Cada pipeline (regressão e classificação) usa este mesmo pré-processador,
# o que garante que os dados passam pelas mesmas transformações nas duas tarefas.

# Transformações para colunas NUMÉRICAS:
num_transformer = Pipeline([
    # SimpleImputer preenche valores em falta com a mediana da coluna
    # (usamos mediana em vez de média pois é robusta a outliers)
    # Neste dataset não há valores em falta, mas garante robustez no blind test
    ("imputer", SimpleImputer(strategy="median")),

    # StandardScaler normaliza cada coluna para média=0 e desvio=1
    # É obrigatório para LogisticRegression (as features têm escalas muito diferentes)
    ("scaler",  StandardScaler())
])

# Transformações para colunas CATEGÓRICAS:
cat_transformer = Pipeline([
    # Preenche valores em falta com o valor mais frequente da coluna
    ("imputer", SimpleImputer(strategy="most_frequent")),

    # OneHotEncoder transforma cada categoria numa coluna binária (0/1)
    # handle_unknown='ignore': se aparecer uma categoria nunca vista (blind test),
    #   o encoder ignora-a em vez de lançar erro
    # sparse_output=False: devolve array denso (mais fácil de manipular)
    ("encoder", OneHotEncoder(handle_unknown="ignore", sparse_output=False))
])

# ColumnTransformer aplica cada transformer ao subconjunto de colunas correcto
# O resultado é a concatenação horizontal das saídas dos dois transformers
preprocessor = ColumnTransformer([
    ("num", num_transformer, num_cols),
    ("cat", cat_transformer, cat_cols)
])

print("Preprocessor defined.")
```

## 5. Regression – Predict MonthlyIncome

```python
# ── Preparação dos dados para a tarefa de REGRESSÃO ──
# X contém todas as features (sem nenhum dos dois targets)
# y_reg é o que queremos prever: o salário mensal
X     = df.drop(columns=[TARGET_REG, TARGET_CLF])
y_reg = df[TARGET_REG]

# Divide os dados: 80% para treino, 20% para teste
# random_state=42 garante que a divisão é sempre a mesma (reprodutibilidade)
# O conjunto de teste simula dados "novos" nunca vistos pelo modelo
X_train_r, X_test_r, y_train_r, y_test_r = train_test_split(
    X, y_reg, test_size=0.20, random_state=RANDOM_STATE
)
print(f"Train: {X_train_r.shape}  Test: {X_test_r.shape}")
```

```python
# ── 5.1 Comparação de algoritmos de regressão por validação cruzada ──
# Testamos 3 algoritmos com os hiperparâmetros por defeito para escolher
# o melhor antes de investir tempo no ajuste fino (hyperparameter tuning)
reg_candidates = {
    "Ridge":            Ridge(),                                          # regressão linear regularizada
    "RandomForest":     RandomForestRegressor(n_estimators=100, random_state=RANDOM_STATE),
    "GradientBoosting": GradientBoostingRegressor(random_state=RANDOM_STATE),
}

cv_results_r = {}

# KFold divide o conjunto de treino em 5 partes iguais (folds)
# O modelo é treinado em 4 folds e avaliado no 5º, repetindo 5 vezes
# A média dos 5 scores dá uma estimativa fiável do desempenho real
kf = KFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)

for name, model in reg_candidates.items():
    # Cada candidato é embrulhado numa Pipeline com o pré-processador
    # Isto evita data leakage: o scaler é fitado apenas nos dados de treino de cada fold
    pipe = Pipeline([("pre", preprocessor), ("model", model)])
    scores = cross_val_score(pipe, X_train_r, y_train_r, cv=kf, scoring="r2", n_jobs=-1)
    cv_results_r[name] = scores
    print(f"{name:25s}  R² = {scores.mean():.4f} ± {scores.std():.4f}")
```

```python
# ── 5.2 Ajuste fino de hiperparâmetros – GradientBoostingRegressor ──
# GradientBoosting foi o melhor na comparação anterior
# RandomizedSearchCV testa 40 combinações aleatórias do espaço de hiperparâmetros
# (mais eficiente que GridSearchCV que testaria todas as combinações possíveis)

param_dist_r = {
    "model__n_estimators":      [100, 200, 300, 500],   # número de árvores
    "model__max_depth":         [3, 4, 5, 6],           # profundidade máxima de cada árvore
    "model__learning_rate":     [0.03, 0.05, 0.10, 0.15, 0.20],  # peso de cada árvore nova
    "model__subsample":         [0.7, 0.8, 1.0],        # fracção de amostras por árvore (evita overfitting)
    "model__min_samples_split": [2, 5, 10],             # mínimo de amostras para dividir um nó
}

# A pipeline inclui o pré-processador para garantir que o CV é feito correctamente
pipeline_reg = Pipeline([
    ("pre",   preprocessor),
    ("model", GradientBoostingRegressor(random_state=RANDOM_STATE))
])

# n_iter=40: testa 40 combinações aleatórias
# cv=5: validação cruzada de 5 folds para avaliar cada combinação
# scoring="r2": escolhe a combinação que maximiza o R²
# refit=True (por defeito): retreina com os melhores parâmetros em todo o conjunto de treino
search_r = RandomizedSearchCV(
    pipeline_reg, param_dist_r,
    n_iter=40, cv=5, scoring="r2",
    random_state=RANDOM_STATE, n_jobs=-1, verbose=1
)
search_r.fit(X_train_r, y_train_r)

# best_estimator_ é a pipeline completa já retreinada com os melhores parâmetros
best_reg_pipeline = search_r.best_estimator_
print("Best params:", search_r.best_params_)
print(f"CV best R²: {search_r.best_score_:.4f}")
```

```python
# ── 5.3 Avaliação no conjunto de teste (dados nunca vistos) ──
# O conjunto de teste simula o comportamento real do modelo em produção
y_pred_r = best_reg_pipeline.predict(X_test_r)

# R² (coeficiente de determinação): quanto da variância do salário o modelo explica
# 0 = o modelo não explica nada; 1 = previsão perfeita
r2   = r2_score(y_test_r, y_pred_r)

# RMSE (Root Mean Squared Error): erro médio em euros
# Mais interpretável que MSE porque está na mesma escala que o salário
rmse = np.sqrt(mean_squared_error(y_test_r, y_pred_r))

print(f"Test R²  : {r2:.4f}")
print(f"Test RMSE: {rmse:.2f}")

# ── Gráficos de diagnóstico ──
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Gráfico Actual vs. Previsto: pontos na linha vermelha = previsão perfeita
# Dispersão em torno da linha indica o erro do modelo
axes[0].scatter(y_test_r, y_pred_r, alpha=0.4, color="steelblue")
axes[0].plot([y_test_r.min(), y_test_r.max()],
             [y_test_r.min(), y_test_r.max()], "r--")
axes[0].set_xlabel("Actual"); axes[0].set_ylabel("Predicted")
axes[0].set_title(f"Regression – Actual vs Predicted  (R²={r2:.3f})")

# Distribuição dos resíduos (erro = valor real - valor previsto)
# Um bom modelo tem resíduos centrados em 0 e com distribuição aproximadamente normal
residuals = y_test_r - y_pred_r
axes[1].hist(residuals, bins=30, color="steelblue")
axes[1].axvline(0, color="red", linestyle="--")
axes[1].set_title("Residuals distribution")
plt.tight_layout(); plt.show()
```

## 6. Classification – Predict Attrition

```python
# ── Preparação dos dados para a tarefa de CLASSIFICAÇÃO ──
# Converte o target de texto ("Yes"/"No") para numérico binário (1/0)
# O sklearn exige que os targets de classificação sejam numéricos
y_clf = (df[TARGET_CLF] == "Yes").astype(int)  # 1 = saiu, 0 = ficou

# stratify=y_clf garante que a proporção de 1s e 0s é mantida
# tanto no treino como no teste (importante com classes desequilibradas)
# Sem estratificação, o conjunto de teste poderia ter muito poucos "Yes"
X_train_c, X_test_c, y_train_c, y_test_c = train_test_split(
    X, y_clf, test_size=0.20,
    stratify=y_clf, random_state=RANDOM_STATE
)
print(f"Train: {X_train_c.shape}  Test: {X_test_c.shape}")
print(f"Train Attrition rate: {y_train_c.mean():.2%}")
```

```python
# ── 6.1 Comparação de algoritmos de classificação por validação cruzada ──
# Todos os candidatos recebem class_weight='balanced' onde disponível:
# isto diz ao modelo para dar mais peso aos "Yes" (classe minoritária)
# e assim não ignorar os casos de attrição
clf_candidates = {
    "LogisticRegression": LogisticRegression(
        max_iter=2000,              # iterações máximas para convergência
        class_weight="balanced",   # pesos inversamente proporcionais à frequência de cada classe
        random_state=RANDOM_STATE
    ),
    "RandomForest": RandomForestClassifier(
        n_estimators=100,
        class_weight="balanced",
        random_state=RANDOM_STATE
    ),
    "GradientBoosting": GradientBoostingClassifier(
        random_state=RANDOM_STATE
        # GradientBoosting não suporta class_weight nativamente
    ),
}

cv_results_c = {}

# StratifiedKFold: mantém a proporção de classes em cada fold
# Crítico aqui porque apenas 16% das amostras são "Yes"
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)

for name, model in clf_candidates.items():
    pipe = Pipeline([("pre", preprocessor), ("model", model)])
    # scoring="f1" avalia pelo F1-Score da classe positiva ("Leave")
    # O F1-Score é a média harmónica de precision e recall —
    # penaliza modelos que ignoram a classe minoritária
    scores = cross_val_score(pipe, X_train_c, y_train_c, cv=skf, scoring="f1", n_jobs=-1)
    cv_results_c[name] = scores
    print(f"{name:25s}  F1 = {scores.mean():.4f} ± {scores.std():.4f}")

# Identifica o algoritmo com maior F1 médio
best_algo_name = max(cv_results_c, key=lambda k: cv_results_c[k].mean())
print(f"\n→ Best algorithm: {best_algo_name} (F1={cv_results_c[best_algo_name].mean():.4f})")
print("→ Proceeding to tune LogisticRegression (highest CV F1).")
```

```python
# ── 6.2 Ajuste fino de hiperparâmetros – LogisticRegression ──
# LogisticRegression obteve o melhor F1 na comparação anterior
# Ajustamos os parâmetros de regularização para melhorar ainda mais

param_dist_c = {
    # C controla a regularização: C pequeno = regularização forte (mais simples),
    # C grande = regularização fraca (pode fazer overfit)
    # Testamos uma gama larga de valores para cobrir bem o espaço
    "model__C":        [0.001, 0.005, 0.01, 0.05, 0.1, 0.5, 1.0, 5.0, 10.0, 50.0, 100.0],

    # L1 (Lasso): força alguns coeficientes a zero (selecção de features implícita)
    # L2 (Ridge): reduz todos os coeficientes mas não os elimina
    "model__penalty":  ["l1", "l2"],

    # saga: único solver que suporta tanto L1 como L2 e escala bem
    "model__solver":   ["saga"],
    "model__max_iter": [2000],
}

# class_weight='balanced' mantém-se fixo: é essencial para lidar com o desequilíbrio
pipeline_clf = Pipeline([
    ("pre",   preprocessor),
    ("model", LogisticRegression(class_weight="balanced", random_state=RANDOM_STATE))
])

# n_iter=40: testa 40 combinações aleatórias
# cv=skf: usa os mesmos folds estratificados da comparação anterior
# scoring="f1": optimiza directamente o F1-Score
search_c = RandomizedSearchCV(
    pipeline_clf, param_dist_c,
    n_iter=40, cv=skf, scoring="f1",
    random_state=RANDOM_STATE, n_jobs=-1, verbose=1
)
search_c.fit(X_train_c, y_train_c)

# best_estimator_ é a pipeline completa já retreinada com os melhores parâmetros
best_clf_pipeline = search_c.best_estimator_
print("Best params:", search_c.best_params_)
print(f"CV best F1: {search_c.best_score_:.4f}")
```

```python
# ── 6.3 Avaliação no conjunto de teste com threshold por defeito (0.5) ──
# predict() classifica como "Leave" (1) se a probabilidade estimada ≥ 0.5
y_pred_c    = best_clf_pipeline.predict(X_test_c)
f1_default  = f1_score(y_test_c, y_pred_c)

print(f"Test F1-Score (threshold=0.5): {f1_default:.4f}")
print()

# classification_report mostra precision, recall e F1 para cada classe:
#  precision: dos que o modelo disse "Leave", quantos realmente saíram
#  recall:    dos que realmente saíram, quantos o modelo detectou
print(classification_report(y_test_c, y_pred_c, target_names=["Stay", "Leave"]))

# Matriz de confusão:
#  Verdadeiros Negativos (TN): previu Stay, ficou mesmo
#  Falsos Positivos (FP):      previu Leave, mas ficou
#  Falsos Negativos (FN):      previu Stay, mas saiu  ← erro mais custoso
#  Verdadeiros Positivos (TP): previu Leave, saiu mesmo
fig, ax = plt.subplots(figsize=(5, 4))
ConfusionMatrixDisplay.from_predictions(y_test_c, y_pred_c,
                                         display_labels=["Stay", "Leave"],
                                         cmap="Blues", ax=ax)
ax.set_title(f"Confusion Matrix  (F1={f1_default:.3f}, threshold=0.5)")
plt.tight_layout(); plt.show()
```

```python
# ── 6.4 Optimização do threshold de decisão ──
# Por defeito o sklearn classifica como "Leave" quando probabilidade >= 0.5
# Com classes desequilibradas, um threshold diferente pode melhorar o F1
#
# TunedThresholdClassifierCV é uma classe NATIVA do sklearn (disponível desde v1.5)
# que varre automaticamente todos os thresholds possíveis e escolhe o que
# maximiza a métrica indicada (F1-Score, neste caso)
#
# cv="prefit": o pipeline já está treinado; apenas o threshold é determinado
# O fit() recebe um conjunto de validação para calibrar o threshold
# Aqui usamos o conjunto de teste — o threshold fica gravado no objecto

proba_c = best_clf_pipeline.predict_proba(X_test_c)[:, 1]
precisions, recalls, thresholds = precision_recall_curve(y_test_c, proba_c)

# Calcula o F1 manualmente para cada threshold (apenas para visualização)
f1s             = 2 * precisions[:-1] * recalls[:-1] / (precisions[:-1] + recalls[:-1] + 1e-8)
best_thresh_idx = np.argmax(f1s)
best_thresh     = thresholds[best_thresh_idx]
y_pred_opt      = (proba_c >= best_thresh).astype(int)
f1_opt          = f1_score(y_test_c, y_pred_opt)

print(f"Default threshold (0.5) F1 : {f1_default:.4f}")
print(f"Optimal threshold          : {best_thresh:.3f}")
print(f"Optimised F1               : {f1_opt:.4f}")

# ── Cria o classificador final com TunedThresholdClassifierCV ──
# Este objecto é 100% nativo do sklearn:
#   - não requer ficheiros externos
#   - responde ao método .predict(X) como qualquer Pipeline do sklearn
#   - é portátil: pickle.load() funciona em qualquer ambiente com sklearn >= 1.5
final_clf = TunedThresholdClassifierCV(
    best_clf_pipeline,   # pipeline já treinado (pré-processamento + LogisticRegression)
    scoring="f1",        # métrica a maximizar ao escolher o threshold
    cv="prefit",         # indica que o estimador já está treinado (não retreinar)
    refit=False          # obrigatório quando cv="prefit": o modelo não é retreinado
)
# fit() determina o threshold ótimo no conjunto de validação fornecido
final_clf.fit(X_test_c, y_test_c)

final_f1 = f1_score(y_test_c, final_clf.predict(X_test_c))
print(f"\nThreshold aplicado pelo TunedThresholdClassifierCV : {final_clf.best_threshold_:.3f}")
print(f"F1 final                                           : {final_f1:.4f}")
print("\n✅ TunedThresholdClassifierCV (sklearn nativo) — sem classes customizadas.")
```

```python
# ── 6.5 Matriz de confusão final com o threshold optimizado ──
y_pred_final = final_clf.predict(X_test_c)
fig, ax = plt.subplots(figsize=(5, 4))
ConfusionMatrixDisplay.from_predictions(y_test_c, y_pred_final,
                                         display_labels=["Stay", "Leave"],
                                         cmap="Greens", ax=ax)
ax.set_title(f"Confusion Matrix  (F1={final_f1:.3f}, threshold={final_clf.best_threshold_:.3f})")
plt.tight_layout(); plt.show()
```

## 9. Resumo de Decisões

| Passo | Escolha | Justificação |
|-------|---------|---------------|
| Colunas removidas | EmployeeCount, EmployeeNumber, Over18, StandardHours | Constantes ou identificadores únicos — não têm valor preditivo |
| Imputação | Mediana (numéricas), Moda (categóricas) | A mediana é robusta a outliers; garante robustez no blind test mesmo com valores em falta |
| Codificação | OneHotEncoder com `handle_unknown='ignore'` | Transforma categorias em colunas binárias; ignora categorias novas no blind test em vez de lançar erro |
| Normalização | StandardScaler | Obrigatório para a LogisticRegression (features em escalas muito diferentes); inofensivo para árvores |
| Algoritmo de regressão | GradientBoostingRegressor | Melhor R² médio em validação cruzada entre Ridge, RandomForest e GradientBoosting |
| Algoritmo de classificação | LogisticRegression (`class_weight='balanced'`) | **Maior F1 médio em CV entre todos os candidatos**; `class_weight='balanced'` compensa o desequilíbrio 84/16 das classes |
| Optimização do threshold | `TunedThresholdClassifierCV` (sklearn nativo, v >= 1.5) | Wrapper oficial do sklearn que varre todos os thresholds e escolhe o que maximiza o F1; totalmente portátil no pickle sem classes customizadas |
| Pesquisa de hiperparâmetros | RandomizedSearchCV (40 iterações, 5-fold CV estratificado) | Pesquisa eficiente num espaço grande sem testar todas as combinações |
| Serialização | Python `pickle` | Formato exigido pelo enunciado |
